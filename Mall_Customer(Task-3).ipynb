```python
df.drop('CustomerID',inplace=True,axis=1)
```

## BIVARIATE ANALYSIS

```python
plt.figure(figsize=(10,5))
for gender in ['Male' , 'Female']:
    plt.scatter(x = 'Age' , y = 'Annual_Income_(k$)' , data = df[df['Genre'] == gender] ,
                s = 200 , alpha = 0.5 , label = gender)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
for gender in ['Male' , 'Female']:
    plt.scatter(x = 'Age' , y = 'Spending_Score_(1-100)' , data = df[df['Genre'] == gender] ,
                s = 200 , alpha = 0.5 , label = gender)
plt.legend()
plt.show()
```
